## **Demo Notebook for using PFF Data**

### 1. Read and Process Data

```python

from ligafooty.data_processor.pff import PFFTrackingDataProcessor

meta_data_json_file=r"data/pff/3844_meta.json"
roster_meta_data_json_file =r"data/pff/3844_roster.json"
raw_data_json_file =r"data/pff/3844.jsonl.bz2"

processor = PFFTrackingDataProcessor()

# Expected runtime around 2-3 mins
tidy_data,gk_shirt_numbers,team_mapping,player_df = processor.process(meta_data_json_file,
                                                            roster_meta_data_json_file ,
                                                            raw_data_json_file,add_velocity=True)
```

### 3. 2D animation for a range of frames

```python
# You need to create a folder to save the videos
import os

# Define the images folder
videos_folder = "videos"

# Check if the folder exists, if not, create it
os.makedirs(videos_folder, exist_ok=True)
```

```python
frame_start =19787
frame_end =19950

# On avg it takes 1 second to process 2-3 frames. So keep in mind to process 50-100 at once. 25 frames make up 1 second of video
processor.animate_frames(tidy_data,gk_df,frame_start,frame_end,mode='voronoi',video_writer="gif",text_font="Arial",save=True) 
```
